## **Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import torch
from datetime import datetime
```

## **Load the Dataset**

```python
path = "../data/processed/preprocessed_energy_data.csv"
data = pd.read_csv(path)
```

```python
#Display first five records
data.head()
```

```python
print(data.head())
```

## **Converting the date column to Date and Time Format**

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
data.head()
```

### **Enhancing the Date column more clearly**

```python
data['hour'] = data['date'].dt.hour
data['month'] = data['date'].dt.month
data['year'] = data['date'].dt.year
```

```python
data.head()
```

```python
window_sizes = [3, 6, 12]
for window in window_sizes:
    data[f'rolling_mean_{window}'] = data['Appliances'].rolling(window=window).mean()
    data[f'rolling_std_{window}'] = data['Appliances'].rolling(window=window).std()
```

```python
data.head()
```

## **Lagged Features**

```python
data['lag_10'] = data['Appliances'].shift(1) # 10 minutes lag
data['lag_30'] = data['Appliances'].shift(3) # 30 minutes lag
data['lag_60'] = data['Appliances'].shift(6) # 60 minutes lag

print(data.head())
```

```python
# autocorrelation plot for Appliances to visualize lag significance
from pandas.plotting import autocorrelation_plot

autocorrelation_plot(data['Appliances'])
plt.show()
```

## **Interaction Features**

```python
# creating interaction features between indoor temperature and humidity
data['in_temp_hum_interaction'] = data['T1'] * data['RH_1']

# creating interaction features between outdoor temperature and humidity
data['out_temp_hum_interaction'] = data['T_out'] * data['RH_out']

# some more interaction features
data['T1_T2_interaction'] = data['T1'] * data['T2']
data['T2_RH_2_interaction'] = data['T2'] * data['RH_2']
data['temp_diff'] = data['T1'] - data['T2']
data['temp_ratio'] = data['T1'] / data['T2']

# Display the interaction features of dataset
print(data[['date','Appliances','in_temp_hum_interaction','out_temp_hum_interaction','T1_T2_interaction','T2_RH_2_interaction','temp_diff','temp_ratio']])
```

## **Domain Specific Features**

```python
data['day'] = data['date'].dt.dayofweek # Monday = 0....., Sunday = 6
data['weekday'] = data['date'].dt.dayofweek < 5
data['weekend'] = data['date'].dt.weekday >= 5  # True for Saturday or Sunday; False for weekdays
```

```python
data.head()
```

### **Encoding Categorical columns**

```python
categorical_columns = data[['day','weekday', 'weekend']]

print(categorical_columns.head())
print(categorical_columns.tail())
```

#### *Since the 'day' column is Ordered, label encoding is most suitable*

#### *Also its better to convert Boolean columns('weekday','weekend') to Integer*

```python
from sklearn.preprocessing import LabelEncoder

# Label encoding
label_encoder = LabelEncoder()
data['day'] = label_encoder.fit_transform(data['day'])

# Converting boolean columns to integer
data['weekday'] = data['weekday'].astype(int)
data['weekend'] = data['weekend'].astype(int)
```

```python
print(data.head())
```

## **Checking whether null values exist**

```python
data.isna().sum()
```

```python
# Generating distribution diagrams for columns which have null values

# Identifying columns with null values
null_columns = data.columns[data.isnull().any()]

# Create distribution diagrams for columns with null values
for col in null_columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(data[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

#### **Since all numerical columns with null values are Skewed-right, its good to use median to fill null values**

```python
for col in null_columns:
    median_val = data[col].median()
    data[col].fillna(median_val, inplace=True)
```

```python
data.isna().sum()
```

## **Feature Selection**

```python
from sklearn.feature_selection import RFE # Feature selection technique (Recursive Feature Elimination)
from sklearn.ensemble import RandomForestRegressor # Machine learning model (Random Forest)
from sklearn.model_selection import train_test_split
```

```python
# Splitting into train and test
X = data.drop(['Appliances', 'date'], axis=1) # Feature set
y = data['Appliances'] # Target variable

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
model = RandomForestRegressor()
```

#### **Tree-Based Models**

```python
# Train a Random Forest model
model.fit(X_train, y_train)

# Get feature importance
importances = model.feature_importances_
feature_importance = pd.DataFrame({'Feature': X.columns, 'Importance': importances}).sort_values(by='Importance', ascending=False)

# Select top 10 important features (adjust based on your needs)
tree_based_features = feature_importance['Feature'].head(10).tolist()
print("Selected features by Tree-Based Model:\n", tree_based_features)
```

#### **Recursive Feature Elimination Method**

```python
# Use RFE to select features
rfe = RFE(estimator=model, n_features_to_select=10)
rfe.fit(X_train, y_train)

# Get the selected features
selected_features = X.columns[rfe.support_]

print("Selected Features by RFE:\n", selected_features)
```

#### **Lasso (L1 Regularization) for Feature Selection**

```python
from sklearn.linear_model import LassoCV
from sklearn.preprocessing import StandardScaler

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Fit Lasso with cross-validation
lasso = LassoCV(cv=5, random_state=42).fit(X_scaled, y)

# Select features where coefficient is not zero
lasso_selected_features = X.columns[lasso.coef_ != 0]

print("Selected Features by Lasso:\n", list(lasso_selected_features))
```

#### **Permutation Feature Importance**

```python
from sklearn.inspection import permutation_importance

result = permutation_importance(model, X_test, y_test, n_repeats=10, random_state=42, n_jobs=-1)

perm_importance_df = pd.DataFrame({
    'Feature': X_test.columns,
    'Importance': result.importances_mean
}).sort_values(by='Importance', ascending=False)

print("Permutation Importance Top Features:\n", perm_importance_df.head(10))

# Optional plot
plt.figure(figsize=(8,6))
plt.barh(perm_importance_df['Feature'][:10][::-1], perm_importance_df['Importance'][:10][::-1])
plt.xlabel("Permutation Importance")
plt.title("Top 10 Permutation Features")
plt.tight_layout()
plt.show()
```

#### **RFECV (RFE with Cross Validation)**

```python
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold

# Use a linear model or any other estimator
estimator = LinearRegression()
cv = KFold(n_splits=5)

rfecv = RFECV(estimator=estimator, step=1, cv=cv, scoring='neg_mean_squared_error')
rfecv.fit(X, y)

rfecv_selected_features = X.columns[rfecv.support_]
print("Selected Features by RFECV:\n", list(rfecv_selected_features))

# Optional plot
plt.figure()
plt.xlabel("Number of features selected")
plt.ylabel("Cross-validation score (neg MSE)")
plt.plot(range(1, len(rfecv.cv_results_['mean_test_score']) + 1), rfecv.cv_results_['mean_test_score'])
plt.title("RFECV Feature Selection")
plt.show()
```

#### **Correlation with Target (Regression-Specific)**

```python
# Compute correlation with target
correlation = X.corrwith(y).sort_values(key=abs, ascending=False)

# Display top correlated features
print("Top features correlated with target:\n", correlation.head(10))

# Select features with absolute correlation above a threshold
threshold = 0.1  # You can tune this
selected_corr_features = correlation[abs(correlation) > threshold].index.tolist()

print("Selected Features by Correlation with Target:\n", selected_corr_features)
```

#### **Combine selected features from all methods**

```python
# Combine selected features from all methods (remove duplicates)
# Convert all feature lists to sets
tree_set = set(tree_based_features)
rfe_set = set(selected_features)
lasso_set = set(lasso_selected_features)
perm_set = set(perm_importance_df['Feature'][:10])  # Top 10 permutation features
rfecv_set = set(rfecv_selected_features)
corr_set = set(selected_corr_features)

# Get the intersection (only features selected by all methods)
combined_features = tree_set & rfe_set & lasso_set & perm_set & rfecv_set & corr_set

print("Combined Selected Features:\n", combined_features)
```

```python
combined_data = data[list(combined_features) + ['Appliances']]
```

```python
print(combined_data)
```

#### **Scale all columns except target variable**

```python
scaled_data = combined_data.drop('Appliances', axis=1)

scaled_features = scaler.fit_transform(scaled_data)
scaled_final_data = pd.DataFrame(scaled_features, columns=scaled_data.columns)

# Concatenate
combined_data = pd.concat([scaled_final_data, combined_data['Appliances']], axis=1)

print(combined_data.head())
```

```python
combined_data.to_csv('..data/processed/combined_data.csv', index=False)
```
